```python
texto = input('Qual termo deseja procurar no arquivo?')

try:
    manipulador = open('arquivo.txt','r', encoding='utf-8')
    for linha in manipulador:
        linha = linha.rstrip()
        if texto in linha:
            print(f'A string foi encontrada!')
            print(linha)
        else:
            print(f'String n√£o encontrada')
except IOError:
    print(f'N√£o foi poss√≠vel abrir o arquivo')
else:
    manipulador.close()
```
