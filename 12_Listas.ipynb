Lista: sequência de valores

sintaxe: nome_lista = [valores]

```python
notas = [5,6,7,8,9,7]
print(notas)
```

```python
n1 = [1,9,6,7,5,8,54]
n2 = [56,69,3,5,8,4,2]
valores = n1 + n2
print(valores)
```

```python
n1 = [1,9,6,7,5,8,54]
n2 = [56,69,3,5,8,4,2]
valores = n1 + n2
print(valores[1])
```

```python
n1 = [1,9,6,7,5,8,54]
n2 = [56,69,3,5,8,4,2]
valores = n1 + n2
valores[1] = 7
print(valores[1])
```

```python
n1 = [1,9,6,7,5,8,54]
n2 = [56,69,3,5,8,4,2]
valores = n1 + n2
print(valores[0:2])
```

```python
n1 = [1,9,6,7,5,8,54]
n2 = [56,69,3,5,8,4,2]
valores = n1 + n2
print(len(valores))
print(sorted(valores))
print(sorted(valores,reverse=True))
print(sum(valores))
print(min(valores))
print(max(valores))
```

```python
valores.append(13)
print(valores)
```

```python
valores.pop()
print(valores)
```

```python
valores.pop(3)
print(valores)
```

```python
valores.insert(3,21)
print(valores)
```

```python
print (54 in valores)
```
