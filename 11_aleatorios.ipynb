```python
import random

print('gerar 5 números aleatórios entre 1 e 50: \n')
for i in range(5):
    n = random.randint(1,50)
    print(f'número gerado: {n}')
```

```python
valor = random.random()
print(f'número gerado: {round(valor * 10,2)}')
```

```python
L = [2,6,5,8,9,3,27,85,546,1863,2549,54,55]
n = random.choice(L)
print(f'Número escolhido: {n}')
```

```python
n = random.sample(L,3)
print(f'Números escolhidos: {n}')
```
