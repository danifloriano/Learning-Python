módulo math

```python
import math

x = 8
y = 100

raiz_quadrada = math.sqrt(x)
print(raiz_quadrada)
print(math.floor(raiz_quadrada))
```

```python
logaritmo = math.log10(y)
print(logaritmo)
print(math.pi)
print(math.factorial(x))
print(x / math.inf)
```
