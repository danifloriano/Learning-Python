```python
x = 5
y = 8
z = 1

def soma_multi(a, b, c = 0):
    if c == 0:
        return a * b
    else:
        return a + b + c

if __name__ == '__main__':
    res = soma_multi(x,y) #vai multiplicar por que c=0
    print(res)

    res = soma_multi(x,y,z) #vai somar por que c não é =0
    print(res)

```
